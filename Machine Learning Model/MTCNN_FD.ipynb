```python
import os
import cv2
from matplotlib import pyplot
from mtcnn import MTCNN
from PIL import Image
import pickle

def get_cropped_image(img):
    IMAGE_SHAPE = (224, 224)
    IMAGE_SHAPE+(3,)
    to_detect = cv2.imread(img)
    detector = MTCNN()
    faces = detector.detect_faces(to_detect)
    for result in faces:
        x,y,w,h = result['box']
        x1,y1 = x+w , y+h
        cropped = to_detect[y:y1, x:x1]
        cropped = cv2.resize(cropped , IMAGE_SHAPE)
        return cropped
```

```python
test_image = "dataset/train/p1/images (1).jpg"
get_cropped_image(test_image)
```
